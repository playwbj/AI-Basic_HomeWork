### 1.实现一周的记账功能

```python
income_str = input("请输入7天的收入以逗号隔开")
```

```python
outlay_str = input("请输入7天的支出，以逗号隔开")
```

```python
income_list = [float(i) for i in income_str.split(',')]
outlay_list = [float(i) for i in outlay_str.split(',')]
daily_dict = [income_list[i]-outlay_list[i] for i in range(len(income_list))]
final_dict = 0
for i in daily_dict:
    final_dict += i;
```

```python
print('每天的收入：',income_list)
print('每天的支出：',outlay_list)
print('每天的结余：',daily_dict)
print('最终的结余：',final_dict)
```

### 2.实现一周的记账功能

```python
balance = 0
end = True
while end:
    ops = input('请输入您要进行的操作：0代表存款，1代表取款，2代表查询余额,3代表退出');
    if ops == '0':
        money = float(input('请输入您要存的金额：'))
        balance+= money
        print('存款成功')
    elif ops == '1':
        money = float(input('请输入您要取的金额：'))
        if balance < money:
            print('取款失败，余额不足')
        else:
            balance -= money
            print('取款成功')
    elif ops == '2':
        print('您的余额：',balance)
    elif ops == '3':
        end = False
        print('退出成功：',balance)
    else:
        print('操作无效')
```
